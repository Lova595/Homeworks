## **Задача 1**

Напишите код, который вычисляет значение следующей математической функции в зависимости от переменной х, где х - вектор

$F(x) = \frac{1}{25}AA^Tx + Bx + 5$

* Создайте матрицы A и B размера 4 на 4, состоящие из случайных целых чисел в интервале от 1 до 10 (используйте соответствующую функцию модуля np.random)
* Создайте вектор х размера 4, состоящие из случайных целых чисел в интервале от 1 до 5
* Перемножение матриц, и матрицы на вектор должно быть матричное.

````
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
````

````
# создадим матрицы 4 на 4
a = np.random.randint(1, 11, (4, 4))
b = np.random.randint(1, 11, (4, 4))
````

````
print(b)
````

````
# Создаем вектор х размера 4, состоящие из случайных целых чисел в интервале от 1 до 5
x = np.random.randint(1,6, (4,))
print(x)
````

````
# вычислим значение математической функции F(x)=1/25*A*At*x+Bx+5
f = 1 / 25 * a @ a.T @ x.reshape(4,) + b @ x.reshape(4,) + 5
````

````
print(f) # выведем полученные заничения
````

````
plt.plot(x, f) # изобразим на графике
plt.show
````

##**Задача 2**

Напишите функцию, которая находит все точки пересечения заданной одномерной математической функции с осью X и отображает их на графике функции красными кругами.

````
# Функция для графика
def f3(x):
  y = x*x*np.sin(x/300) + 300*x
  return y
````

````
# Функция Python, которая выводит график y
# y - вектор значений функции "y = ..."
# a - начальная точка диапазона чисел для массива x
# b - конечная точка диапазона чисел для массива x
def plot_vect(y, a, b):
  x = np.arange(a, b)  # создаем массив значений по x
  # Отображаем график
  plt.plot(x, y)
  plt.grid(True)  # отрисовываем сетку на графике
  plt.show()

# Функция, которая выводит график Python функции f
def plot_func(f, a, b):
  x = np.arange(a, b)  # создаем массив значений по x
  # Вычисляем функцию в этих точках
  y = np.array([f(i) for i in x])

  plot_vect(y, a, b)  # рисуем график

  return y   # возвращаем массив значений y
````

````
# Выводим график функции
y = plot_func(f3, -5000, 5000)
````

````
#  Вывести все точки пересечения заданной одномерной математической функции с осью X и отобразить их на графике функции красными кругами
def find_zero_onx(f, a, b):
  x = np.arange(a, b)    # создаем массив значений х
  y = np.array([f(i) for i in x]) # вычисление функции в этих точках

  # Пустой лист для точкек пересечений
  zero_onx = []

  # Проходим по диапазону массива значений функции
  for i in range(len(y)): # если точка равна 0, то добавляем
    if abs(y[i]) < 2500:  # особенностью питона является невозможность напрямую сравнивать точки с 0, поэтому используем значение по модулю и меньше заданного диапазона
      zero_onx.append(i)

  # Выводим график функции y
  plt.plot(x, y)
  plt.grid(True)
  # Выводим нулевые значения
  plt.plot ([x[i] for i in zero_onx],
            [y[i] for i in zero_onx],
            'ro')
````

````
#Выводим функцию с пересечениями
plt.figure(figsize=(10,10))
find_zero_onx(f3, -5000, 5000)
````

## **Задача 3**

Напишите функцию, которая получает на вход пару матриц и проверяет закон коммутативности матричного умножения, то есть свойства $A*B=B*A$, возвращая ```1``` в случае подтверждения и ```0``` в противном случае. Поставьте ```1000``` экспериментов для случайных целочисленных матриц размера ```(5,5)``` элементов, которые принадлежат отрезку ```[-5,5]```, и найдите количество успешных срабатываний.

````
# создаем рандомно две матрицы
m_1 = np.random.randint(-5, 5, (5,5))
m_2 = np.random.randint(-5, 5, (5,5))
````

````
# сщздаем функцию, которая проверяет закон коммутативности матричного умножения
def check_com(matrix_1, matrix_2):
  if (matrix_1 @ matrix_2 == matrix_2 @ matrix_1).all(): # если все элементы равны - вернется 1.
    return 1
  else:
    return 0

# ставим 1000 экспериментов для случайных целочисленных матриц
def experiment():
  count = 0
  for i in range(1000):
    matr_1 = np.random.randint(-5, 5, (5,5))
    matr_2 = np.random.randint(-5, 5, (5,5))
    rezult = check_com(matr_1, matr_2)
    if rezult == 1:
      count += rezult
      print('Успешная проверка, количество успешных срабатываний: ', rezult)
  print('Эксперимент не подтвердил гипотезу, количество успешных срабатываний: ', rezult)
````

````
check_com(m_1, m_2)
````

````
experiment()
````
